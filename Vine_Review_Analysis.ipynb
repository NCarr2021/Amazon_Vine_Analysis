```python
# Dependencies

import pandas as pd
```

```python
# Create a DataFrame for vine_table
 
df = pd.read_csv("../BigData/Resources/vine_table.csv", low_memory=False)
df
# results = 1741095
```

```python
# Filter DF and create new DF for total votes

total_votes_df = df.loc[(df["total_votes"] >= 20)]
total_votes_df
# results = 35312
```

```python
# Filter DF to create new DF for percentage

total_percentvotes_df = total_votes_df.loc[(total_votes_df["helpful_votes"] / total_votes_df["total_votes"] >= 0.5)]
total_percentvotes_df
# results = 31830
```

```python
# Filter DF to create new DF for paid vine

paid_vines_df = total_percentvotes_df.loc[(total_percentvotes_df["vine"] == "Y")]
paid_vines_df 
# results = 285
```

```python
# Filter DF to create new DF for unpaid vine

unpaid_vines_df = total_percentvotes_df.loc[(total_percentvotes_df["vine"] == "N")]
unpaid_vines_df 
# results = 31545
```

```python
# Paid vine
# Total number of reviews (285)

len(paid_vines_df)
```

```python
# Paid vine
# Number of 5 star reviews(163)

fivestar_paid_df = paid_vines_df.loc[(df["star_rating"] == 5)]
len(fivestar_paid_df)
```

```python
# Paid vine
# Percentage of 5 star reviews

pd1 = paid_vines_df.count()
fv1 = fivestar_paid_df.count()
paid_pct = (fv1 / pd1 * 100).map("{:,.1f}%".format)
paid_pct
```

```python
# Unpaid vine
# Total number of reviews (31545)

len(unpaid_vines_df)
```

```python
# Unpaid vine
# Number of 5 star reviews (14614)

fivestar_unpaid_df = unpaid_vines_df.loc[(df["star_rating"] == 5)]
len(fivestar_unpaid_df)
```

```python
# Unpaid vine
# Percentage of 5 star reviews

pd1 = unpaid_vines_df.count()
fv1 = fivestar_unpaid_df.count()
unpaid_pct = (fv1 / pd1 * 100).map("{:,.1f}%".format)
unpaid_pct
```
